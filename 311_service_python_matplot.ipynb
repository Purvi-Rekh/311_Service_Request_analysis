```python
#Display top 10 complaint types
complaint_grp = df['Unique Key'].groupby(df['Complaint Type']).count()
print(type(complaint_grp))
print(complaint_grp.sort_values(ascending=False)[0:9])
```

```python
#Visualize major compaint types and their count
print(complaint_grp)
```

```python
#plot a bar graph of count vs Complaint type
%matplotlib inline
group_by_compain =df['Unique Key'].groupby(df['Complaint Type'])
cnt = group_by_compain.count()
cnt.plot.bar()
```
